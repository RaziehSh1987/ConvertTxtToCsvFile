**convert .txt to .csv file**

```python
# convert text file in C:\Users\18195\Desktop\AnimationTransform.txt to the csv file
# the csv file is used to train the model

# load the text file
import csv

# read the text file
with open('/content/gdrive/MyDrive/BishopThesis2/AnimationTransform.txt', 'r') as f:
    reader = csv.reader(f,delimiter=' ')
    data = list(reader)
    # create the csv file and seprate each element  with space delimeter into different columns
    with open('/content/gdrive/MyDrive/BishopThesis2/AnimationTransform11.csv', 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)
```
